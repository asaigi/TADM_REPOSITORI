```python
import numpy as np
from numpy import genfromtxt
```

```python

valors = np.genfromtxt("USAhousing.csv",delimiter=",", usecols=[11], skip_header=1)
price=np.array(valors)
```

```python
valors2 = np.genfromtxt("USAhousing.csv",delimiter=",", usecols=[19], skip_header=1)
councilstreet=np.array(valors2)
```

```python
council_price=np.column_stack((councilstreet, price))
print(council_price)
```

```python

np.isnan(council_price).any(axis=1)#Â¿existe algun valor nulo en la matriz?
```

```python
council_price=council_price[~np.isnan(council_price).any(axis=1)]

np.isnan(council_price).any(axis=1)#ahora ya no existe
```

```python
print(council_price.shape)
```

```python
np.sort(council_price, axis = 0) #ordenar por el nÃºmero de distritos
```

```python
filtro=np.where(council_price[:,0]==20)
council_price[filtro]
```

```python
district20=council_price[filtro]

np.mean(district20[:,1])
#Observam que en definitiva, coincideix i per tant, correspon que la mitjana del districte 20 es igual a 16208333.333333334 
#o en el mateix cas, com es veu al grafic inicial, 1.6e+07.
```

```python
#Â¿cuÃ¡l es el precio mÃ¡s elevado de una casa en USA?
max_price=np.max(council_price[:,1])
max_price20=np.max(district20[:,1])
a = np.zeros(shape=2)
#guardaremos la informaciÃ³n total en la columna 1 y la del distrito 20 en la segunda.
a[0]=max_price
a[1]=max_price20

print("el precio maximo del distrito",a[1],"es significativamente menor al total ", a[0])
print(a)
```

```python
#Â¿cuÃ¡l es el precio mÃ¡s bajo de una casa en USA?
min_price=np.min(council_price[:,1])
min_price20=np.min(district20[:,1])

b = np.zeros(shape=2)
#guardaremos la informaciÃ³n total en la columna 1 y la del distrito 20 en la segunda.
b[0]=min_price
b[1]=min_price20
print("observamos que el precio min del distrito es significativamente mayor al total (de", round(b[0])," a",round(b[1]),")") 
```

```python
#Â¿cuÃ¡l es la media del precio de una casa en USA?
mean_price=np.mean(council_price[:,1])
mean_price20=np.mean(district20[:,1])
c = np.zeros(shape=2)
#guardaremos la informaciÃ³n total en la columna 1 y la del distrito 20 en la segunda.
c[0]=mean_price
c[1]=mean_price20
print("observamos que el precio medio del distrito",round(c[1],2), "es un poco mayor al total ",round(c[0],2),"(casi el doble)") 
```

```python
#Â¿cuÃ¡l es la mediana del precio de una casa en USA?
median_price=np.median(council_price[:,1])
median_price20=np.median(district20[:,1])
d = np.zeros(shape=2)
#guardaremos la informaciÃ³n total en la columna 1 y la del distrito 20 en la segunda.
d[0]=median_price
d[1]=median_price20
print("observamos que el punto medio del distrito",round(d[1])," es relativamente menor al total",round(d[0])) 
```
